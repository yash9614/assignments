```python
#36. Find the frequency of each character in a string.

def character_frequency(input_string):
    
    #initialise an empty dict to store character frequencies
    freq_dict = {}
    for char in input_string:
        if char in freq_dict:
            freq_dict[char] += 1
        else:
            freq_dict[char] = 1
    return freq_dict


# Test the function
input_str = "Hello, World!"
result = character_frequency(input_str)

# Print the character frequencies
for char, count in result.items():
    print(f"'{char}': {count}")
```

```python

# 42.  Encode a string using a Caesar cipher.

def caesar_cipher(text, shift):
    encoded_text = ""
    
    for char in text:
        if char.isalpha():
            # Determine whether the character is uppercase or lowercase
            is_upper = char.isupper()
            char = char.lower()  # Convert to lowercase for easier manipulation
            
            # Apply the Caesar cipher shift
            shifted_char = chr(((ord(char) - ord('a') + shift) % 26) + ord('a'))
            
            # Convert back to uppercase if the original character was uppercase
            if is_upper:
                shifted_char = shifted_char.upper()
            
            encoded_text += shifted_char
        else:
            # If the character is not alphabetic, leave it unchanged
            encoded_text += char
    
    return encoded_text

# Test the function
input_text = "Hello, World!"
shift_value = 3

encoded_text = caesar_cipher(input_text, shift_value)
print(encoded_text)
```

```python

# 43. Find the most frequent word in a string.


def most_frequent_word(input_string):
    # Tokenize the string into words (split by whitespace and remove punctuation)
    words = input_string.split()
    word_frequency = {}

    # Remove punctuation from each word and count the frequency
    for word in words:
        # Remove common punctuation characters
        word = word.strip('.,!?":;')
        word = word.lower()  # Convert to lowercase for case-insensitive counting
        if word:
            word_frequency[word] = word_frequency.get(word, 0) + 1

    if not word_frequency:
        return None  # No words found

    # Find the word with the highest frequency
    most_frequent = max(word_frequency, key=word_frequency.get)
    return most_frequent, word_frequency[most_frequent]

# Test the function
input_text = "This is a test. This test is a test, a test of word frequency."

most_frequent, frequency = most_frequent_word(input_text)
print(f"The most frequent word is '{most_frequent}' with a frequency of {frequency} times.")
```

```python
# 45. Find all unique words in a string.



import re

def unique_words(input_string):
    # Tokenize the string into words (split by whitespace and remove punctuation)
    words = re.findall(r'\b\w+\b', input_string.lower())

    # Use a set to store unique words
    unique_word_set = set(words)

    return list(unique_word_set)

# Test the function
input_text = "This is a test. This test is a test, a test of unique words."

unique_word_list = unique_words(input_text)
print(unique_word_list)
```

```python

# 46. Count the number of syllables in a string.


import re

def count_syllables(word):
    # Remove common suffixes that don't contribute to syllables
    word = re.sub(r'e$', '', word, flags=re.IGNORECASE)
    word = re.sub(r'ed$', '', word, flags=re.IGNORECASE)
    word = re.sub(r'es$', '', word, flags=re.IGNORECASE)
    
    # Count the number of vowel sequences
    syllables = re.findall(r'[aeiouy]+', word, flags=re.IGNORECASE)
    
    return len(syllables)

# Test the function
word1 = "apple"
word2 = "elephant"
word3 = "programming"

syllables1 = count_syllables(word1)
syllables2 = count_syllables(word2)
syllables3 = count_syllables(word3)

print(f"The number of syllables in '{word1}' is {syllables1}.")
print(f"The number of syllables in '{word2}' is {syllables2}.")
print(f"The number of syllables in '{word3}' is {syllables3}.")
```

```python
# 48. Remove the nth word from a string.


def remove_nth_word(input_string, n):
    # Split the input string into words
    words = input_string.split()

    # Check if the given index is valid
    if 1 <= n <= len(words):
        # Remove the nth word (subtract 1 from n to adjust for 0-based indexing)
        del words[n - 1]

        # Join the remaining words into a single string
        result_string = ' '.join(words)
        return result_string
    else:
        # Invalid index, return the original string
        return input_string

# Test the function
input_str = "This is a sample sentence to demonstrate removing the nth word."
n = 3  # Remove the third word

new_str = remove_nth_word(input_str, n)
print(new_str)
```

```python
# 49. Insert a word at the nth position in a string.


def insert_word_at_nth_position(input_string, word_to_insert, n):
    # Split the input string into words
    words = input_string.split()

    # Check if the given index is valid
    if 0 <= n <= len(words):
        # Insert the new word at the nth position
        words.insert(n, word_to_insert)

        # Join the words back into a single string
        result_string = ' '.join(words)
        return result_string
    else:
        # Invalid index, return the original string
        return input_string

# Test the function
input_str = "This is a sample sentence to demonstrate inserting a word."
word_to_insert = "new"
n = 5  # Insert the word at the fifth position (0-based index)

new_str = insert_word_at_nth_position(input_str, word_to_insert, n)
print(new_str)
```

```python
#50. Convert a CSV string to a list of lists.

import csv
from io import StringIO

def csv_string_to_list(csv_string):
    # Use StringIO to treat the string as a file-like object
    csv_file = StringIO(csv_string)
    
    # Initialize a CSV reader
    reader = csv.reader(csv_file)
    
    # Read the CSV data into a list of lists
    csv_list = [row for row in reader]
    
    return csv_list

# Test the function
csv_data = "Name,Age,Location\nJohn,30,New York\nAlice,25,Los Angeles\nBob,35,Chicago"

result = csv_string_to_list(csv_data)
for row in result:
    print(row)
```

```python
def is_subsequence(subseq , seq):
    subseq_length = len(subseq)
    seq_length = len(seq)
    
    if subseq_length > seq_length:
        return False
    
    subseq_index = 0
    for item in seq:
        if subseq[subseq_index] == item:
            subseq_index +=1
            if subseq_index == subseq_length:
                return True
    return False


main_list = [1, 2, 3, 4, 5, 6, 7, 8, 9]
subsequence_list = [2, 4, 6]

# Check if subsequence_list is a subsequence of main_list
result = is_subsequence(subsequence_list, main_list)

if result:
    print(True)
else:
    print(False)
```

```python
# 39. Rotate a tuple to the left by `n` positions.


def rotate_tuple(tup, n):
    if len(tup) == 0:
        return tup
    
    n %= len(tup)
    return tup[n:] + tup[:n]

#Example Usage:

original_tuple = (2,1,2,4,9)

n = 2

rotated_tuple = rotate_tuple(original_tuple, n)

rotated_tuple
```

```python
# 40. Rotate a tuple to the right by `n` positions.


def rotate_tuple_right(tup, n):
    if len(tup) == 0:
        return tup
    
    n %= len(tup)
    return tup[-n:] + tup[:-n]

original_tuple = (1, 2, 3, 4, 5)
n = 2
rotated_tuple = rotate_tuple_right(original_tuple, n)
print("Original Tuple:", original_tuple)
print(f"Rotated {n} Positions to the Right:", rotated_tuple)
```

```python
# 49. Check if a tuple contains another tuple as a subsequence.

def is_sub_sequence(subseq, seq):
    subseq_len = len(subseq)
    seq_len = len(seq)
    
    if subseq_len > seq_len:
        return False
    
    subseq_index = 0
    for item in seq:
        if subseq[subseq_index] == item:
            subseq_index += 1
            if subseq_index == subseq_len:
                return True
            
    return False

# Example usage:
sequence = (1, 2, 3, 4, 5, 6, 7, 8)
subsequence = (3, 5, 7)

result = is_sub_sequence(subsequence, sequence)
print("Is subsequence:", result)
```
