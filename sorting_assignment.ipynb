```python
'''
Problem 1.Given an array of n numbers, give an algorithm which gives the element appearing maximum
number of times?
'''
def find_max_frequency_element(arr):
    frequency_map = {}

    for num in arr:
        if num in frequency_map:
            frequency_map[num] += 1
        else:
            frequency_map[num] = 1

    max_frequency = 0
    max_frequency_element = None

    for num, frequency in frequency_map.items():
        if frequency > max_frequency:
            max_frequency = frequency
            max_frequency_element = num

    return max_frequency_element

# Example usage:
arr = [1, 2, 3, 2, 4, 2, 5, 1, 2, 4, 2]
result = find_max_frequency_element(arr)
print("Element appearing maximum number of times:", result)

```

```python
'''
Problem 2 : We are given a list of n-1 integers and these integers are in the range of 1 to n . There are no
duplicates in the list. One of the integers is missing in the list. Give an algorithm to find that element Ex:
[1,2,4,6,3,7,8] 5 is the missing num.
'''
def find_missing_element(nums):
    n = len(nums) + 1  # n is the size of the original list

    expected_sum = n * (n + 1) // 2
    actual_sum = sum(nums)

    missing_element = expected_sum - actual_sum
    return missing_element

# Example usage:
nums = [1, 2, 4, 6, 3, 7, 8]
result = find_missing_element(nums)
print("Missing element:", result)
```

```python
'''
Problem 3 : Given an array of n positive numbers. All numbers occurs even number of times except 1 which
occurs odd number of times. Find that number in O(n) time and O(1) space. Ex: [1,2,3,2,3,1,3]. 3 is repeats odd
times.

'''
def find_odd_occurrence(arr):
    result = 0

    for num in arr:
        result ^= num

    return result

# Example usage:
arr = [1, 2, 3, 2, 3, 1, 3]
result = find_odd_occurrence(arr)
print("Number occurring odd number of times:", result)
```

```python
'''
Problem 4 : Given an array of n elements. Find two elements in the array such that their sum is equal to given
element K.
'''
def find_pair_with_sum(arr, K):
    seen_elements = set()

    for num in arr:
        complement = K - num
        if complement in seen_elements:
            return num, complement
        seen_elements.add(num)

    return None

# Example usage:
arr = [1, 2, 4, 5, 6]
K = 10
result = find_pair_with_sum(arr, K)

if result:
    print(f"Pair with sum {K}: {result[0]} and {result[1]}")
else:
    print(f"No pair found with sum {K}")
```

```python
'''
Problem 5 : Given an array of both positive and negative numbers, find two numbers such that their sum is
closest to 0. Ex: [ 1 ,60 ,-10, 70, -80,85]. Ans : -80,85.
'''
def closest_sum_pair(arr):
    arr.sort()

    left, right = 0, len(arr) - 1
    closest_sum = float('inf')
    pair = None

    while left < right:
        current_sum = arr[left] + arr[right]

        # Update closest_sum and pair if the current sum is closer to zero
        if abs(current_sum) < abs(closest_sum):
            closest_sum = current_sum
            pair = (arr[left], arr[right])

        # Move pointers based on the sign of the current sum
        if current_sum < 0:
            left += 1
        else:
            right -= 1

    return pair

# Example usage:
arr = [1, 60, -10, 70, -80, 85]
result = closest_sum_pair(arr)

print("Closest sum pair:", result)
```

```python
'''
Problem 6 : Given an array of n elements . Find three elements such that their sum is equal to the given
number.
'''
def find_three_elements_with_sum(arr, target):
    n = len(arr)
    arr.sort()

    for i in range(n - 2):
        left, right = i + 1, n - 1

        while left < right:
            current_sum = arr[i] + arr[left] + arr[right]

            if current_sum == target:
                return arr[i], arr[left], arr[right]
            elif current_sum < target:
                left += 1
            else:
                right -= 1

    return None

# Example usage:
arr = [1, 4, 2, 8, 3, 9, 5]
target = 15
result = find_three_elements_with_sum(arr, target)

if result:
    print(f"Three elements with sum {target}: {result}")
else:
    print(f"No such three elements found.")
```

```python
'''
Problem 7 : Given an array of n elements . Find three elements i, j, k in the array such that
i * i + j * j = k*k.
'''
def find_pythagorean_triplet(arr):
    n = len(arr)
    arr_squares = [num * num for num in arr]

    arr_squares_set = set(arr_squares)

    for i in range(n - 1):
        for j in range(i + 1, n):
            sum_of_squares = arr_squares[i] + arr_squares[j]

            if sum_of_squares in arr_squares_set:
                return arr[i], arr[j], int(sum_of_squares ** 0.5)

    return None

# Example usage:
arr = [3, 1, 4, 6, 5]
result = find_pythagorean_triplet(arr)

if result:
    print(f"Pythagorean triplet found: {result}")
else:
    print("No such triplet found.")
```

```python
'''
Problem 8 : An element is a majority if it appears more than n/2 times. Give an algorithm takes an array of n
element as argument and identifies a majority (if it exists).
'''
def find_majority_element(arr):
    candidate = None
    count = 0

    for num in arr:
        if count == 0:
            candidate = num
            count = 1
        elif candidate == num:
            count += 1
        else:
            count -= 1

    # Check if the candidate is a majority element
    count = 0
    for num in arr:
        if num == candidate:
            count += 1

    if count > len(arr) // 2:
        return candidate
    else:
        return None

# Example usage:
arr = [2, 2, 3, 2, 5, 2, 2]
result = find_majority_element(arr)

if result is not None:
    print(f"Majority element: {result}")
else:
    print("No majority element found.")
```

```python
'''
Problem 9 : Given n × n matrix, and in each row all 1’s are followed by 0’s. Find the row with the maximum
number of 0’s.
'''
def find_max_zeros_row(matrix):
    max_zeros_row_index = -1
    max_zeros_count = -1

    n = len(matrix)

    for i in range(n):
        # Count zeros in the current row
        zeros_count = matrix[i].count(0)

        # Update max_zeros_row_index if the current row has more zeros
        if zeros_count > max_zeros_count:
            max_zeros_count = zeros_count
            max_zeros_row_index = i

    return max_zeros_row_index

# Example usage:
matrix = [
    [1, 1, 1, 0],
    [1, 1, 0, 0],
    [1, 0, 0, 0],
    [1, 1, 1, 1]
]

result = find_max_zeros_row(matrix)

if result != -1:
    print(f"Row with maximum number of zeros: {result}")
else:
    print("No such row found.")
```
